```python
from sklearn.linear_model import LinearRegression
import pandas
import numpy
import matplotlib.pyplot as mat_plot
import seaborn
```

```python
data = pandas.read_csv("Salary.csv")
```

```python
print(data.head())
```

```python
mat_plot.scatter(data["YearsExperience"],data["Salary"])
mat_plot.xlabel("Years of Experience")
mat_plot.ylabel("Salary")
mat_plot.show()
```

```python
def predict(X,w,b):
  return w*X+b;
```

```python
def loss(y,y_hat):
  return (1/len(y)) * numpy.sum((y-y_hat)**2)
```

```python
def gradient(X,w,b,alpha,y,y_pred):
  new_w = w - alpha * (1/len(y)) * numpy.sum(X * (y_pred - y))
  new_b = b - alpha * (1/len(y)) * numpy.sum(y_pred - y)
  return new_w,new_b
```

```python
w = numpy.random.randn(1)[0]
b = numpy.random.randn(1)[0]
epoch = 100
X = data["YearsExperience"]
y = data["Salary"]
loss_list = []
alpha = 0.01
```

```python
for i in range(epoch):
  y_pred = predict(X,w,b)
  loss_ = loss(y,y_pred)
  loss_list.append(loss_)
  w,b = gradient(X,w,b,alpha,y,y_pred)
```

```python
mat_plot.scatter(X,y)
mat_plot.plot(X,predict(X,w,b),color="red")
mat_plot.show()
```

```python
mat_plot.plot(loss_list)
mat_plot.show()
```
