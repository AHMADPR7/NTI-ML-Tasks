```python
import pandas as pd
import matplotlib.pyplot as plt
import kagglehub
import os
```

```python
# Download latest version
path = kagglehub.dataset_download("alirezachahardoli/customer-data-clustring")
print("Path to dataset files:", path)
```

```python
df = pd.read_csv(os.path.join(path, 'Customer-Data - 2.csv'))
df.head()
```

```python
df.info()
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.info()
df.isnull().sum()
```

```python
df.describe()
```

```python
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
feature_columns = df.columns[1:-1]
X = df.loc[:, feature_columns].values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

eps = 2.0
min_samples = 10

dbscan = DBSCAN(eps=eps, min_samples=min_samples)
cluster_labels = dbscan.fit_predict(X_scaled)

n_clusters = len(set(cluster_labels) - {-1})
n_noise = int((cluster_labels == -1).sum())

df_clustered = df.copy()
df_clustered['cluster'] = cluster_labels

print(f"DBSCAN parameters: eps={eps}, min_samples={min_samples}")
print(f"Clusters found (excluding noise): {n_clusters}")
print(f"Noise points: {n_noise} / {len(cluster_labels)}")
```

### Why accuracy was zero
This dataset does not contain a supervised target label for customer segments. The last column, `TENURE`, is a feature, not a ground-truth class, so comparing it directly with DBSCAN cluster IDs gives a misleading accuracy score. For DBSCAN, use cluster counts, noise ratio, silhouette score, and plots instead.

```python
cluster_summary = (
    pd.Series(cluster_labels, name='count')
    .value_counts()
    .sort_index()
    .rename_axis('cluster')
    .reset_index()
)

print('Cluster counts:')
print(cluster_summary.to_string(index=False))

mask = cluster_labels != -1
if n_clusters > 1 and mask.sum() > n_clusters:
    silhouette = silhouette_score(X_scaled[mask], cluster_labels[mask])
    print(f'\nSilhouette score (excluding noise): {silhouette:.3f}')
else:
    silhouette = None
    print('\nSilhouette score is not defined because DBSCAN found fewer than 2 non-noise clusters.')

print('\nTENURE distribution by cluster (for interpretation only):')
print(pd.crosstab(df_clustered['cluster'], df_clustered['TENURE']))
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

unique_labels = sorted(np.unique(cluster_labels))
n_non_noise_labels = max(len([label for label in unique_labels if label != -1]), 1)
palette = plt.cm.tab10(np.linspace(0, 1, n_non_noise_labels))

color_index = 0
for label in unique_labels:
    label_mask = cluster_labels == label
    if label == -1:
        color = 'black'
        label_name = 'Noise'
    else:
        color = palette[color_index % len(palette)]
        label_name = f'Cluster {label}'
        color_index += 1

    axes[0].scatter(
        X_pca[label_mask, 0],
        X_pca[label_mask, 1],
        s=18,
        alpha=0.65,
        c=[color],
        label=label_name
    )

explained_variance = pca.explained_variance_ratio_.sum() * 100
axes[0].set_title(f'DBSCAN clusters in 2D PCA space ({explained_variance:.1f}% variance)')
axes[0].set_xlabel('Principal component 1')
axes[0].set_ylabel('Principal component 2')
axes[0].legend()

cluster_counts = pd.Series(cluster_labels).value_counts().sort_index()
bar_colors = ['black' if label == -1 else 'steelblue' for label in cluster_counts.index]
axes[1].bar(cluster_counts.index.astype(str), cluster_counts.values, color=bar_colors)
axes[1].set_title('Points per cluster')
axes[1].set_xlabel('Cluster label')
axes[1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
eps_values = np.arange(1.0, 2.6, 0.2)
min_samples_values = [5, 10, 15, 20, 30]

records = []
for eps_candidate in eps_values:
    eps_candidate = round(float(eps_candidate), 1)
    for min_samples_candidate in min_samples_values:
        sweep_labels = DBSCAN(eps=eps_candidate, min_samples=min_samples_candidate).fit_predict(X_scaled)
        n_clusters_candidate = len(set(sweep_labels) - {-1})
        noise_ratio = float((sweep_labels == -1).mean())
        score = np.nan

        valid_mask = sweep_labels != -1
        if n_clusters_candidate > 1 and valid_mask.sum() > n_clusters_candidate:
            score = silhouette_score(X_scaled[valid_mask], sweep_labels[valid_mask])

        records.append({
            'eps': eps_candidate,
            'min_samples': min_samples_candidate,
            'n_clusters': n_clusters_candidate,
            'noise_ratio': noise_ratio,
            'silhouette': score,
        })

sweep_df = pd.DataFrame(records)
print('Best configurations from the sweep:')
print(
    sweep_df.sort_values(['silhouette', 'n_clusters'], ascending=[False, False])
    .head(10)
    .to_string(index=False)
)

cluster_pivot = sweep_df.pivot(index='min_samples', columns='eps', values='n_clusters')
silhouette_pivot = sweep_df.pivot(index='min_samples', columns='eps', values='silhouette')

fig, axes = plt.subplots(1, 2, figsize=(16, 5))

im0 = axes[0].imshow(cluster_pivot.values, aspect='auto', cmap='viridis')
axes[0].set_title('Number of clusters across settings')
axes[0].set_xlabel('eps')
axes[0].set_ylabel('min_samples')
axes[0].set_xticks(range(len(cluster_pivot.columns)))
axes[0].set_xticklabels([f'{value:.1f}' for value in cluster_pivot.columns])
axes[0].set_yticks(range(len(cluster_pivot.index)))
axes[0].set_yticklabels(cluster_pivot.index)
fig.colorbar(im0, ax=axes[0])

silhouette_values = np.nan_to_num(silhouette_pivot.values, nan=-1)
im1 = axes[1].imshow(silhouette_values, aspect='auto', cmap='magma', vmin=-1, vmax=max(0.6, np.nanmax(silhouette_pivot.values)))
axes[1].set_title('Silhouette score (-1 means undefined)')
axes[1].set_xlabel('eps')
axes[1].set_ylabel('min_samples')
axes[1].set_xticks(range(len(silhouette_pivot.columns)))
axes[1].set_xticklabels([f'{value:.1f}' for value in silhouette_pivot.columns])
axes[1].set_yticks(range(len(silhouette_pivot.index)))
axes[1].set_yticklabels(silhouette_pivot.index)
fig.colorbar(im1, ax=axes[1])

current_eps = round(float(eps), 1)
current_x = list(cluster_pivot.columns).index(current_eps)
current_y = list(cluster_pivot.index).index(min_samples)
for ax in axes:
    ax.scatter(current_x, current_y, marker='x', s=120, color='cyan', linewidths=2)

plt.tight_layout()
plt.show()
```
